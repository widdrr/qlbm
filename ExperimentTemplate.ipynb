```python
%load_ext autoreload
%autoreload 2

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime
import os
import qlbm
from qlbm.initial_conditions import *
from qlbm.visualization import *
```

```python
# ── Lattice ──────────────────────────────────────────────────────────
EXPERIMENT_NAME = "MyExperiment"
NUM_ITERATIONS  = 50

# D1Q3 example — replace as needed
links           = [[-1], [0], [1]]
weights         = [1/6, 2/3, 1/6]
speed_of_sound  = 1 / np.sqrt(3)

# ── Grid & initial conditions ─────────────────────────────────────────
sites           = (32,)
initial_density = get_gaussian_initial_distribution(sites[0], center=sites[0]//4, width=2.0)
velocity_field  = get_uniform_velocity_field(sites[0], velocity=0.2)

# ── Boundary conditions (None = periodic) ─────────────────────────────
boundary_conditions = None

# ── Config (add more tuples for multi-phase simulations) ──────────────
config = [(NUM_ITERATIONS, velocity_field, links, weights, speed_of_sound, boundary_conditions)]
```

```python
timestamp  = datetime.now().strftime("%Y%m%d_%H%M%S")
dirname    = f"experiments/{EXPERIMENT_NAME}_{timestamp}"
os.makedirs(dirname, exist_ok=True)

classical_csv = f"{dirname}/classical.csv"
quantum_csv   = f"{dirname}/quantum.csv"

qlbm.simulate_flow_classical(initial_density, config, classical_csv)
qlbm.simulate_flow(initial_density, config, quantum_csv)
```

```python
save_simulation_snapshots_1d(
    classical_csv, quantum_csv,
    iterations=[0, NUM_ITERATIONS//3, 2*NUM_ITERATIONS//3, NUM_ITERATIONS],
    output_path=f"{dirname}/snapshots.png",
)

qlbm.save_rmse_comparison(classical_csv, quantum_csv, sites, f"{dirname}/rmse.png")
```
